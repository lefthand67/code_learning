## Formatting text
Type without spaces between special marks
- ** bold ** , 
- _ italic _ , 
- ** _ bold and italic _ ** , _ ** bold and italic ** _ .


## Paragraphs
- _soft breaks_ - make a new line; type two spaces in the end of the line;
- _hard breaks_ - make a new paragraph, include a blank line in between paragraphs.


## Headers

There are 6 header levels in markdown. h1, h2, h3, h4, h5 and h6. h1 is the largest, h6 is the smallest. To write a header, include # before the text for h1 header, ## before the text for h2 header, ### before the text for h3 header, and so on.

**NOTE:** In Jupyter notebooks, you must leave a space between the group of # signs and the text. This is not required in markdown in general, but rather it's a Jupyter notebook quirk.

**NOTE:** In Jupyter notebooks, header 5 (h5) is shown as both bold and italic.

**NOTE:** In Jupyter notebooks, header 5 (h5) and header 6 (h6) are displayed identically.

_Example:_ #### this is header 4 
#### this is header 4

## Links

Wrap the text for the link in [ ] and wrap the URL in ( ).

**NOTE:** Remember to include http:// or https:// for absolute links, otherwise the links will be considered relative.

this is a link to the [Marxist Internet Archive](https://www.marxists.org/).

## Images

Adding images is very similar to adding links. The only difference is that the square brackets are prefaced with an exclamation point.

![Lenin is on reading the "Pravda"](https://img0.liveinternet.ru/images/attach/b/0/14434/14434656_Izmenenie_razmera_1720711.jpg)

_Lenin is on reading the "Pravda"_

**NOTE:** When displaying an image in markdown, the text between the square brackets (e.g. [Lenin is on reading the "Pravda"]) becomes the value assigned to the alt attribute of the image - screen readers use the value of this attribute to describe the image if it's not downloaded on the page. You can leave it blank but it is better to fill it with the descriptive text.

## Blockquotes

Blockquotes are used to draw attention to quotes from other sources. Include a "greater than" sign (>) before the text you want to quote.

> If the writer of these lines has succeeded in providing some material for clarifying these problems, he may regard his labours as not having been fruitless. _Lenin V.I._, **The Development of Capitalism in Russia**, “The Mission of Capitalism” (1899)

**NOTE:** You can combine various markdown formatting to achieve exactly what you need as shown in the example above.

## Lists

We have two types of lists: ordered and unordered.

### Ordered lists
In ordered lists each list item starts with a number. What's interesting about markdown _in Jupyter_ notebooks is that you don't have to include the numbers yourself. You can start each list item with 1, for example, and when the markdown is displayed, the correct numbers will be filled in. This feature makes it very easy to reorder items without having to edit all those numbers.
