## Основные понятия и примеры проверки статистических гипотез

**Проверка гипотез** — это важный этап работы над продуктом. Для проверки гипотез нужно освоить специальный математический аппарат и ввести несколько определений. В первую очередь надо определить две гипотезы, с которыми будем работать: 

**Нулевая гипотеза** ($H_0$) — утверждение о параметре генеральной совокупности (параметрах генеральных совокупностей) или распределении, которое необходимо проверить.

**Альтернативная гипотеза** ($H_A$) — утверждение, противоположное нулевой гипотезе. Альтернативная гипотеза выдвигается, но не проверяется.

**Важно:** Нулевая гипотеза всегда проверяется на определенном уровне значимости. Уровень значимости  — вероятность отвергнуть верную нулевую гипотезу.

```python
# Давайте придумаем несколько примеров задач и гипотез о них

# Что-нибудь сгенерируем из нормального? (np.random.normal)

xN = np.random.normal(180, 5, 1000)

# И биномиального? (np.random.binomial)

xB = np.random.binomial(100, 0.3, 1000)
```

```python
# Сейчас давайте поговорим о том, что значит отвергунть гипотезу.

# Пусть у нас есть некоторый эксперимент, статистика которого распределена биномиально. С число экспериментов равное 16 и вероятностью 0.5
# В этот раз мы будем пользоваться генератором случайных чисел из библиотеки scipy

F = stats.binom(16,0.5)
x = np.linspace(0, 16, 17)

# pmf - probabilyty mass function
plt.bar(x, F.pmf(x), align='center')
plt.show()

# Давайте подумаем причем тут вообще статистики с их распределением!:)

# ...
# Статистика -- это функция от выборки. А поскольку выборка это тоже набор реализаций
# случайных величин, то и статистика -- это случайная величина. А значит, эта случайная 
# величина имеет распределение. Обычно, статистики строятся так, чтобы можно было при 
# истинности нулевой гипотезы различить значения, свидетельствующие о нулевой гипотезе или 
# скорее против нее.
```

```python
# Какие значения статистик можно считать аномальными?

x_critical = F.ppf(0.95)

plt.bar(x, F.pmf(x), align='center')
plt.bar(x_critical, F.pmf(x_critical), align='center', color='red')
plt.xlim(-0.5,16.5)
plt.show()
```

## Проверка гипотез

### Рассмотрим несколько критериев:

1. **Z-критерий** -- проверка равества среднего выборки некоторому числу при известной дисперсии. Статистика $T = \frac{\overline{x}-\mu}{\sigma/\sqrt{n}} \sim N(0,1)$

```python
# Пример 1

# Продавец утверждает, что средний вес пачки чая составляет 100 г, а разброс 1 г. Из партии извлечена выборка и взвешена.
# Вес каждой пачки - [98, 104, 97, 97, 101, 100, 99, 101, 99, 98]. Не противоречит ли это утверждению продавца?
# Доверительная вероятность 99%. Вес пачек распределен нормально. 

# Определим функцию, которая по данным считает счатистику

alpha = 0.01
x = np.array([98, 104, 97, 97, 101, 100, 99, 101, 99, 98])
T = np.sqrt(x.shape[0]) * (np.mean(x) - 100)/1

# Воспользуемся функцией библиотеки scipy, которая считает квантили (.ppf()) для нормального распределения stats.norm

n01 = stats.norm(0,1)

# Сделаем выводы

print(f"Квантиль {alpha/2:.3f} и {1-alpha/2:.3f} N(0,1). Границы критической области: {n01.ppf(alpha/2):.3f}, {n01.ppf(1-alpha/2):.3f}")
print(f"Значение статистики {T:.3f}")


to_plot = np.linspace(-3,3,100)
plt.plot(to_plot, n01.pdf(to_plot))

left_crit = np.linspace(-3, n01.ppf(alpha/2))
right_crit = np.linspace(n01.ppf(1-alpha/2), 3)
plt.fill_between(left_crit, n01.pdf(left_crit), alpha=0.4, color="red")
plt.fill_between(right_crit, n01.pdf(right_crit), alpha=0.4, color="red")
plt.scatter(T, 0, color="green")
```

```python
# Пример 2

# Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр
# d=10 мм. Используя односторонний критерий с alpha = 0,05 , проверить эту гипотезу, если в выборке из n=16 шариков средний диаметр
# оказался равным 10,3 мм, а разброс sigma = 1. 

alpha = 0.05
N = 16
md = 10
sigma = 1
T = np.sqrt(N)*(10.3 - md)/1

print(f"Квантиль {1-alpha:.3f} N(0,1). Граница критической области: {n01.ppf(1-alpha):.3f}")
print(f"Значение статистики {T:.3f}")


to_plot = np.linspace(-3,3,100)
plt.plot(to_plot, n01.pdf(to_plot))

right_crit = np.linspace(n01.ppf(1-alpha), 3)
plt.fill_between(right_crit, n01.pdf(right_crit), alpha=0.4, color="red")
plt.scatter(T, 0, color="green")
```

X -- распр нормально

2. Тест стьюдента (одновыборочный, есть много других) -- проверка равества среднего выборки некоторому числу при неизвестной дисперсии. Статистика $T = \frac{\overline{x}-\mu}{s/\sqrt{n}} \sim St(n-1)$

```python
# Пример 3

# Продавец утверждает, что средний вес пачки чая составляет 100 г, а РАЗБРОС НЕИЗВЕСТЕН. Из партии извлечена выборка и взвешена.
# Вес каждой пачки - [98, 104, 97, 97, 101, 100, 99, 101, 99, 98]. Не противоречит ли это утверждению продавца?
# Доверительная вероятность 99%. Вес пачек распределен нормально. 

# Определим функцию, которая по данным считает счатистику

x = np.array([98, 104, 97, 97, 101, 100, 99, 101, 99, 98])
mu = 100
N = x.shape[0]
T = np.sqrt(N) * (np.mean(x) - mu)/np.std(x, ddof=1)
alpha=0.05

# Воспользуемся функцией библиотеки scipy, которая считает квантили (.ppf())
# для распределения стьюдента stats.t с n-1 степенью свободы (как думаете почему так?)

st = stats.t(N-1)

# Сделаем выводы

print(f"Квантиль {alpha/2:.3f} и {1-alpha/2:.3f} st({N-1}). Границы критической области: {st.ppf(alpha/2):.3f}, {st.ppf(1-alpha/2):.3f}")
print(f"Значение статистики {T:.3f}")


to_plot = np.linspace(-3,3,100)
plt.plot(to_plot, n01.pdf(to_plot))

left_crit = np.linspace(-3, st.ppf(alpha/2))
right_crit = np.linspace(st.ppf(1-alpha/2), 3)
plt.fill_between(left_crit, st.pdf(left_crit), alpha=0.4, color="red")
plt.fill_between(right_crit, st.pdf(right_crit), alpha=0.4, color="red")
plt.scatter(T, 0, color="green")

# Воспользуемся встроенной функцией t-критерия -- stats.ttest_1samp

stats.ttest_1samp(x, popmean=mu, ) # alternative='two-sided'
```

```python
# Пример 4

# Давайте приудмаем задачу сами, понимая, что нужно, чтобы применить данный критерий:)
# Требование: все элементы выборки сгенерированы из одного нормального распределения, при этом 
# элементы выборки получены независимо.

# Мы можем придумать какие-нибудь случайные параметры для распределения генератора
# потом посмотреть на график, прикинуть математическое ожидание, и затем проверить тестом
# после чего подсмотрим истинное значение мат ожидания в сгенерированном числе, посмотрим насколько
# мы окажемся правы
import numpy as np
import random
from scipy import stats
from matplotlib import pyplot as plt

mu = random.randint(1,20)
sigma = random.randint(5,30)
N = 100

my_norm = stats.norm(mu, np.sqrt(sigma))

sample = my_norm.rvs(N)
# .rvs дает нам случайную выборку заданного распределения

plt.hist(sample)
```

## Прикладное применение проверки статистических гипотез

Тут нам понадобится еще один критерий (и в практике, наверное, самый популярный) -- t-test для двух выборок.
Статистика выглядит в нем следующим образом:

А распределена она $\sim St(df)$, где df:

```python
# Пример 5

# У вас есть интернет магазин. Было решено ввести новый функционал -- после добавления товара в корзину -- рядом с ней пишется какой процент вы
# вы экономите при помощи скидок, акций и программы лояльности. Для проверки того, что этот функционал положительно влияет на продажи, было решено
# провести АВ тест. Он длился 3 недели, у вас есть данные по суммарным продажам в каждый день. Вам нужно проверить есть ли положительное влияние
A_sample = np.array([207076, 191868, 176711, 207404, 196964, 183714, 185031, 197014,
             217361, 192396, 203864, 222403, 197208, 198619, 206228, 213269,
             202670, 190073, 181998, 204068, 216976])
B_sample = np.array([228437, 245876, 164642, 257656, 208687, 230056, 220714, 220283,
             244674, 202487, 237079, 268262, 192123, 223988, 206281, 232277,
             224813, 208811, 192765, 216265, 219305])

# Будем это делать при помощи критерия stats.ttest_ind
# ttest_ind означает t test для двух независимых выборок.
print(stats.ttest_ind(A_sample, B_sample)) # alternative two-sided, equal_var=True
print(stats.ttest_ind(A_sample, B_sample, alternative='less', ))# alternative less, equal_var=True
print(stats.ttest_ind(A_sample, B_sample, alternative='less', equal_var=False))# alternative less, equal_var=False
```

Можем ли мы в данном случае говорить о равенстве дисперсий? Можно посмотреть на дисперсию визуально, и также посмотреть точечные оценки. Либо, мы можем исходить из каких-то априорных предположений о задаче, но лучше все-таки проверить.

```python
plt.hist(A_sample)
plt.hist(B_sample, alpha=0.5)
```

Мы можем даже провести тест [Фишера](https://ru.wikipedia.org/wiki/F-%D1%82%D0%B5%D1%81%D1%82) на равенство дисперсий.

```python
F = A_sample.var(ddof=1)/B_sample.var(ddof=1)
alpha = 0.05
fdist = stats.f(A_sample.shape[0]-1, B_sample.shape[0]-1,)
print(f"Квантиль {alpha/2:.3f} и {1-alpha/2:.3f} F({A_sample.shape[0]-1}, {B_sample.shape[0]-1}). Границы критической области: {fdist.ppf(alpha/2):.3f}, {fdist.ppf(1-alpha/2):.3f}")
print(f"Значение статистики {F:.3f}")

to_plot = np.linspace(-1,5,100)
plt.plot(to_plot, fdist.pdf(to_plot))

left_crit = np.linspace(-1, fdist.ppf(alpha/2))
right_crit = np.linspace(fdist.ppf(1-alpha/2), 5)
plt.fill_between(left_crit, fdist.pdf(left_crit), alpha=0.4, color="red")
plt.fill_between(right_crit, fdist.pdf(right_crit), alpha=0.4, color="red")
plt.scatter(F, 0, color="green")
```

Этот тест довольно сильно чувствителен к "ненормальности" исходных данных. Поэтому можно на практике использовать тест [Bartlet](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.bartlett.html) или [Levene](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.levene.html#scipy.stats.levene)

```python
# Пример 6

# У вас есть некоторая задача машинного обучения и есть модель машинного обучения, которая решает на данный момент эту задачу. Ваша команда разработала
# новую модель и хочет проверить, что она действительно работает лучше старой. Для этого вы провели АВ тест на отложенных 30 выборках.
# Можно ли утверждать, что новая модель работает лучше старой?
A_sample = np.array([0.58630958, 0.8950967 , 0.67386928, 0.74961669, 0.6644486 ,
             0.87480831, 0.73489688, 0.78344588, 0.96830322, 0.66953365,
             0.93903279, 0.88088772, 0.81794925, 0.89634539, 0.79638788,
             0.63935939, 0.87870204, 0.64540828, 0.77433847, 0.66062788,
             0.47753131, 0.84713575, 0.87831679, 0.85374387, 0.96198082,
             0.89377488, 0.76343049, 0.98538711, 0.94910174, 0.68040054])
B_sample = np.array([0.91299363, 0.76004446, 0.80211475, 0.77121029, 0.90329942,
             0.9346002 , 0.99947112, 0.76803393, 0.99024783, 0.97694625,
             0.84115031, 0.76867202, 0.82324768, 0.61827477, 0.50599651,
             0.94877043, 0.64548139, 0.91042593, 0.74235915, 0.8684409 ,
             0.88263016, 0.84313834, 0.75965375, 0.72364365, 0.65799765,
             0.85031338, 0.59552783, 0.97647643, 0.89606074, 0.9279395 ])

print(stats.ttest_ind(A_sample, B_sample, alternative='less'))# alternative less, equal_var=True
```
