# 1. Теория вероятностей

## Основные понятия
В бытовом понимании вероятность — это степень уверенности в том, что то или иное событие произойдет. Эту вероятность можно измерять числом от 0 до 1. 0 — событие гарантированно не случится, 1 — событие точно произойдет. Но как посчитать вероятность точно? Можно пойти опытным путем и проводить длинную серию экспериментов. Например, 1000 раз подбросить монету и каждый раз отмечать, в какой доле случаев выпал орел. Ниже представлен график такой серии, из которого отчетливо видно: чем больше проведено экспериментов, тем ближе доля к 1/2.  
<p></p>
<div><img src="https://cs.sberbank-school.ru/image/full/full/resize/42385092-ec3d-11ea-a571-0242966aaeec" alt="нарастающее среднее" style="width: 600px; margin-right: 75%"></div>

Но в более общих случаях проделывать эксперимент 1000 раз ради столь простого результата неэффективно. И можно было в нашем примере догадаться, что для симметричной монеты вероятность выпадения орла и решки всякий раз одинаковая, а значит, ответ будет 1/2. 

Теперь формализуем ключевые понятия теории вероятностей, чтобы можно было и в этом, и в более сложных примерах проще и обоснованнее делать суждения о том, чему равна вероятность тех или иных событий. Начнем с понятия элементарного исхода. 

**Элементарный исход** — просто один из возможных результатов какого-то случайного эксперимента. Совокупность всех элементарных исходов называется вероятностным пространством.

Например, рассмотрим бросок игральной кости. Может выпасть только одно из чисел: от одного из шести. Результат выпадения каждого из шести чисел и есть элементарный исход (обозначают обычно греческой буквой "**омега**" с соответствующим индексом). 
<p></p>
<div><img src="https://cs.sberbank-school.ru/image/full/full/resize/7b7c363e-ec3d-11ea-b977-02429a728377" alt="кубики" style="width: 400px; margin-right: 75%"></div>

Набор элементарных исходов называется **событием**. Например, событие может состоять в том, что на игральной кости выпало нечетное число очков (интуитивно понятно, что вероятность такого события равна 1/2). Важно выделить два частных случая.

- **Достоверное событие** — событие, которое произойдет точно. В нашем примере это событие заключается в том, что выпало хоть что-то.
- **Невозможное событие** — в нашем примере невозможно, что не выпало ничего. Вероятность невозможного события равна 0. 

События могут быть разными и сильно зависят от природы задачи. 

Например, Google и Apple рассматривают возможность покупки компаний А, В и С. Могут быть, например, события:

- Apple купил A и B, а Google не купил ничего;
- Apple купил A, Google купил B и С;

и многие другие события.


## Определение вероятности
Нам бы хотелось научиться считать вероятности элементарных исходов и событий. 

Начнем с простейшего случая. Иногда у нас есть возможность предполагать, что исходы в данном эксперименте равновероятны. Например, если мы подбросим симметричную монету, то нет оснований полагать, что орел будет выпадать с большей вероятностью, чем решка или наоборот. Соответственно, логично предположить, что вероятности отдельных элементарных исходов одинаковы. В этом случае вероятность каждого элементарного исхода будет вычисляться как `1/N`, где `N` — это суммарное количество элементарных исходов. 

Закрепим на устных примерах. Вероятность выпадения орла и решки, конечно же, равна `1/2`, поскольку всего у нас два элементарных исхода, они равновероятны (вероятности их наступления равны), соответственно вероятность каждого `1/2`. Тоже самое можно сказать про случай броска игрального кубика. Всего у нас возможно выпадение шести граней, кубик симметричный, соответственно вероятность выпадения той или иной грани одинакова, т.е. все элементарные исходы равновероятны, и получаем что вероятность выпадения какого-то конкретного числа, например `3`, это просто `1/6`.

Но как же тогда подсчитать вероятность более сложных событий? Например, вероятность выпадения нечетного числа на кубике. Развивая логику прошлых абзацев: для произвольного `события А`, состоящего из `N` элементарных исходов, вероятность будет вычисляться как `n/N`, где `n` — число благоприятных элементарных исходов, т.е. подходящих нам, а `N` — это суммарное число элементарных исходов.

## Действия над вероятностями
В общем случае нам бы хотелось научиться конструировать из простых событий более сложные и научиться пересчитывать вероятности наиболее простым образом, чтобы не рассчитывать вероятности сложных событий с нуля. Для этого введем некоторые операции над событиями, которые нам будут нужны в дальнейшем. 

В некотором смысле такие операции эквивалентны некоторым операциям над числами. Например: **сложение, вычитание, умножение** и так далее. Свои операции есть и над событиями. Пусть у нас есть событие `А` (некоторый набор элементарных исходов) и событие `В` (тоже некоторый набор элементарных исходов). Введем сначала формальные определения и потом рассмотрим, как ими пользоваться на практических примерах.

1. **Пересечение событий А и В** — событие, которое заключается в том, что произошло событие А `И` событие В.
<p></p>
<div><img src="https://cs.sberbank-school.ru/image/1000/auto/upsize/0dc247d0-9126-11eb-bd9a-0242ac190003" alt="пересечение" style="width: 200px; margin-right: 75%"></div>

На картинке это можно действительно иллюстрировать, как их пересечение.

Если у событий `А` и `В` пересечений нет (то есть у них нет общих элементарных исходов), то пересечением А и В будет `пустое событие` и вероятность его будет равна `0`. 

Если события `А` и `В` пересекаются (то есть у них есть некоторые общие элементарные исходы), то вероятность пересечения ни по какой универсальный формуле рассчитать нельзя, ее нужно аккуратно подсчитывать, рассматривая общие элементарные исходы. В дальнейшем мы рассмотрим на практике, как это сделать.

<p></p>
<div></div>

2. **Объединение событий А и В** — событие, которое заключается в том, что произошло  событие А `ИЛИ` событие В. То есть хотя бы одно из двух.
<p></p>
<div><img src="https://cs.sberbank-school.ru/image/1000/auto/upsize/0dc547a0-9126-11eb-bfbf-0242ac190002" alt="объединение" style="width: 200px; margin-right: 75%"></div>

На картинке это иллюстрируется соответствующим интуитивным образом (подходят **все исходы** `А` и **все исходы** `В`). 

Если событие `А` и событие `В` не пересекаются (такие события называются **несовместными**), то вероятность их объединения - сумма вероятностей, поскольку мы просто объединили все элементарные события исхода события `А` и все элементарные исходы события `В`.

Если `А` и `В` _пересекаются_, то есть у них есть некоторые общие элементарные исходы, то, рассчитав вероятность объединения по предыдущей формуле, мы допустим серьезную ошибку. Мы учтем те элементарные исходы, которые принадлежат обоим событиям дважды. Поэтому для того, чтобы итоговая формула стала верной 

> из суммы вероятности `А` и `В` необходимо вычесть вероятность их пересечения. 

Такая формула будет работать в общем случае.

<p></p>

## Независимые события
Рассмотрим важный частный случай, который часто встречается на практике и сильно облегчает вычисления. Это случай **независимых событий**.

#### Независимые события

> События `А` и `В` называются независимыми, если наступление одного из них не влияет на другое.

Формально `А` и `В` считаются независимыми, если вероятность пересечения `А` и `В` равна произведению вероятности произведения события `А` отдельно и события `В` отдельно. 
<p></p>
<div></div>

Примерами могут служить двукратные бросания кубика. Очевидно, когда мы бросаем кубик второй раз, результат этого броска никак не зависит от результата первого броска. Или, например, результат сделки с партнером и погода в Милане — совершенно никак не связанные события, которые никак не могут друг на друга повлиять. Однако, когда мы считаем два события независимыми, нужно быть очень аккуратными и учитывать все факторы, которые могут быть существенными. Например, если сделка с партнером происходит в Милане, то плохая погода может существенно отразиться на настроении партнера и, соответственно, результатах сделки. 

## Случайные величины
Теперь, когда мы познакомились с событиями, их вероятностями, с элементарными исходами, с операциями над событиями, настало время перейти к центральному понятию теории вероятностей: к **случайным величинам**. 

Что же такое случайная величина? Это просто _некоторая численная характеристика результата случайного испытания, случайного эксперимента_. Например, рост случайного прохожего, который заранее предсказать мы, конечно же, не можем, оценка пользователям приложения на AppStore от 1 до 5 звездочек или стоимость сделки, которая также заранее неизвестна. 

Формально мы _каждому элементарному исходу ставим в соответствие некоторое число_, которое характеризует этот элементарный исход. Например, сумму очков при подбрасывании двух игральных костей.
<p></p>

## Распределения случайных величин
Конечно, не хотелось бы для каждого случайного процесса строить полноценную таблицу. Это не очень удобный формат для восприятия данных. Хотелось бы уметь по любому распределению случайной величины строить удобные и понятные характеристики, которые выражались бы в виде одного-двух чисел. И такие характеристики существуют: это **математическое ожидание** (грубо говоря, среднее значение величины) и **дисперсия** (характерный разброс величины). Более подробно эти понятия будут рассмотрены во втором модуле курса. 

Оказывается, помимо дискретных распределений, которые можно записать в виде таблички, есть еще **непрерывные распределения**. Это зависит от природы случайной величины. Оказывается, бывают: 

- **Дискретные случайные величины** — это те, которые принимают несколько фиксированных значений. Например, оценка пользователем приложений на AppStore. Это случайная величина, которая соответствует либо одной звездочке, либо двум звездочкам, либо трем, либо четырем, либо пяти. То есть всего пять фиксированных значений. В общем случае их может быть сколь угодно много, но главное - счетное количество.

- **Непрерывные случайные величины**, которые принимают произвольные значения или произвольные значения из некоторого диапазона. Например, рост прохожего, стоимость сделки — они могут равняться произвольному действительному числу. Вопрос: а как же задавать непрерывные случайные величины, если множество их значений бесконечно? Выписать их в таблицу не представляется возможным. В этом случае строится график функции `f(x)`, где `х` — это значения случайной величины, а `f(x)` — это так называемая **плотность вероятности**.

Функция `f(x)*dx`, где `f(x)` - плотность вероятности, а дельта х - диапазон изменения х. Например, для определить плотность в диапазоне от 25 до 30: f(25)x5, т.к. dx = 30-25 = 5.

Рассмотрим теперь несколько основных случайных распределений, которые наиболее часто описывают реальные жизненные ситуации.

### 1. Равномерное распределение.
Равномерное распределение соответствует случаю, когда все значения случайной величины равновероятны. 

<div><img src="https://cs.sberbank-school.ru/image/full/full/resize/d2734566-9130-11eb-951f-0242ac190003" alt="Пример. Равномерное распределение" style="width: 300px; margin-right: 75%"></div>

Мы с ним уже сталкивались. Например, количество очков на кубике — это равномерное дискретное распределение. Действительно, случайная величина дискретна, она принимает шесть случайных значений — 1, 2, 3, 4, 5 и 6. И вероятность каждого значения 1/6. Соответственно, если мы изобразим это распределение, это будет просто шесть столбиков равной высоты. Все значения случайной величины равновероятны, стало быть, это равномерное распределение. Этот случай соответствует дискретным случайным величинам, но и непрерывные распределения могут быть равномерными (в качестве упражнения постарайтесь придумать пример).

### 2. Нормальное распределение. 
Это непрерывное распределение имеет характерный вид колокола, как показано на графике. У него есть пик в центре, и дальше он симметрично спадает до нуля по мере удаления от пика. 

<div><img src="https://cs.sberbank-school.ru/image/full/full/resize/d54124c0-9130-11eb-b5f5-0242ac190002" alt="Пример. Нормальное распределение" style="width: 400px; margin-right: 75%"></div>

Такое распределение очень часто встречается в жизни, например, разброс при стрельбе. В среднем стрелок всегда попадает в центр, то есть наиболее вероятное значение именно в центре, однако периодически пули попадают левее и правее, и чем дальше попадание пули от центра, тем реже это встречается, потому что в среднем стрелок целится в центр. 

Другой пример — объем инвестиций в стартап. Скажем, в среднем объем инвестиций составляет 10 млн долларов, однако в некоторые стартапы инвестируют 9 млн, в некоторые 11 млн, в некоторые 1 млн, в некоторые 100 млн, но две последних ситуации встречаются гораздо реже, чем среднее значение 10 млн. Например, практически нормальным будет распределение количества сотрудников в различных компаниях. Есть очень большие компании, есть очень маленькие компании, но большинство компаний будет находиться в окрестности некоторого среднего значения. И мы тоже увидим некоторое характерное колоколообразное распределение, очень близкое к нормальному.

### 3. Биномиальное распределение. 
Это дискретное распределение получается в результате так называемой **схемы Бернулли**, имеет похожую на случай нормального распределения форму колокола. Подробнее о схеме Бернулли и примерах можно узнать в видео, которые сюда не подгрузить.

<div><img src="https://cs.sberbank-school.ru/image/full/full/resize/d925483c-9130-11eb-949d-0242ac190002" alt="Пример. Нормальное распределение" style="width: 400px; margin-right: 75%"></div>

Формула подсчета k успешных сделок из N **независимых** сделок:

- Вероятность успешной сделки: `P(успеха одной сделки) = р`.   
- Вероятность НЕуспешной сделки: `Р(неуспеха одной сделки) = q = 1 - p`.

Задача: посчитать вероятность свершения сценария, в котором k успешных сделок из N.
- `P(k сделок из N окажется удачными)` - ?.

`x(i)` принимает два значения:
- 1, если сделка успешная,
- 0, если сделка неуспешная.

Следовательно, X успешных сделок будет состоять из суммы x(i), в которых учитываться будут только успешные сделки.

Для вывода формулы используется формула подсчета сочетаний, см. тет. по алгебер №1 (Киселев):

`P(k успешных сделок) = C(из N по k) * p**k * (1-p)**(N-k)`

Сочетания: `n! / k!\*(N-k)!`

Для решения задачи необходимо задать: количество сделок `N`, количество успешных сделок `k`, вероятность успешной сделки `p`.

_См. тет. по алгебре, с.310_
