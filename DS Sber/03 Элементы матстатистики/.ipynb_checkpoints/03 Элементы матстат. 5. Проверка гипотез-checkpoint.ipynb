### –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞

### t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
–¢–µ–ø–µ—Ä—å –≤–≤–µ–¥–µ–º –ø–æ–Ω—è—Ç–∏–µ t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏.

–ü—É—Å—Ç—å –∏—Å—Ç–∏–Ω–Ω–æ–µ –∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ —Ä–∞–≤–Ω—è–µ—Ç—Å—è `Œº`, –∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ, –ø–æ—Å—á–∏—Ç–∞–Ω–Ω–æ–µ –ø–æ –≤—ã–±–æ—Ä–∫–µ –∏–∑ `n` –æ–±—ä–µ–∫—Ç–æ–≤, —Ä–∞–≤–Ω—è–µ—Ç—Å—è `s`. –¢–æ–≥–¥–∞ –≤–µ—Ä–Ω–æ —Å–ª–µ–¥—É—é—â–µ–µ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ: 

> —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ `t = x—Å—Ä - Œº / s/n**1/2` –∏–º–µ–µ—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –°—Ç—å—é–¥–µ–Ω—Ç–∞ —Å `k=n-1` —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã. 

–ü–æ—è—Å–Ω–∏–º. –ï—Å–ª–∏ –±—ã –º—ã –∑–Ω–∞–ª–∏ –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ `Œº`, –≤–∑—è–ª–∏ –º–Ω–æ–≥–æ —Ä–∞–∑–Ω—ã—Ö –≤—ã–±–æ—Ä–æ–∫ –∏–∑ `n` –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏ –¥–ª—è –∫–∞–∂–¥–æ–π –≤—ã–±–æ—Ä–∫–∏ –ø–æ—Å—á–∏—Ç–∞–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ `t` (–ø–æ–ª—É—á–∏–ª–∏ –±—ã –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π `t`), —Ç–æ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –∑–Ω–∞—á–µ–Ω–∏–π `t` –∏–º–µ–ª–∞ –±—ã —Ñ–æ—Ä–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞ c `k=n-1` —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã. 

–ö–∞–∫ –º—ã –ø–æ–º–Ω–∏–º, —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –°—Ç—å—é–¥–µ–Ω—Ç–∞ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ 0, –ø—Ä–∏—á–µ–º —á–µ–º –¥–∞–ª—å—à–µ –æ—Ç 0 –æ—Ç–∫–ª–æ–Ω—è–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏, —Ç–µ–º –º–µ–Ω—å—à–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å —Ç–∞–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ –¥–∞–Ω–Ω—ã—Ö.

–¢–µ–ø–µ—Ä—å –≤—ã—Ä–∞–∑–∏–º –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ `Œº` (–¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –º—ã —Ö–æ—Ç–∏–º –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª) –∏–∑ —Ñ–æ—Ä–º—É–ª—ã –¥–ª—è t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏:

`Œº = x—Å—Ä - t * s/(n**1/2)`

–ò–∑ —Ñ–æ—Ä–º—É–ª—ã –≤–∏–¥–Ω–æ, —á—Ç–æ —á–µ–º –±–ª–∏–∂–µ —á–∏—Å–ª–æ t –∫ 0, —Ç–µ–º –±–ª–∏–∂–µ `x—Å—Ä` –ø—Ä–∏–±–ª–∏–∂–∞–µ—Ç—Å—è –∫ –∏—Å—Ç–∏–Ω–Ω–æ–º—É –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é `Œº`.

–î–æ–≥–æ–≤–æ—Ä–∏–º—Å—è, —á—Ç–æ –±—É–¥–µ–º –∏—Å–∫–∞—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –Ω–∞ —É—Ä–æ–≤–Ω–µ –¥–æ–≤–µ—Ä–∏—è 95%. –¢–æ–≥–¥–∞, –µ—Å–ª–∏ –º—ã –≤–æ–∑—å–º–µ–º –∫–≤–∞–Ω—Ç–∏–ª–∏ 0,025 –∏ 0,975 –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞ (–∏ –æ—Ç–±—Ä–æ—Å–∏–º 2,5% —Å–∞–º—ã—Ö –º–∞–ª–æ–≤–µ—Ä–æ—è—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª–µ–≤–∞ –∏ 2,5% —Å–ø—Ä–∞–≤–∞ ‚Äî —Å–º. —Ä–∏—Å.), —Ç–æ –º—ã ¬´–∑–∞–∂–º–µ–º¬ª –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ Œº —Å–ª–µ–≤–∞ –∏ —Å–ø—Ä–∞–≤–∞:

`x—Å—Ä - t* * s/(n**1/2) < Œº < x—Å—Ä + t* * s/(n**1/2)`

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º—ã –ø–æ—Å—Ç—Ä–æ–∏–ª–∏ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª —Å —É—Ä–æ–≤–Ω–µ–º –¥–æ–≤–µ—Ä–∏—è 95%.

### –í–µ–ª–∏—á–∏–Ω–∞ t*

–ó–∞–º–µ—Ç–∏–º, —á—Ç–æ `t*` ‚Äî –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–≤–∞–Ω—Ç–∏–ª–µ–π ‚Äî —Ç–∞–±–ª–∏—á–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –º—ã –º–æ–∂–µ–º –≤—ã—è—Å–Ω–∏—Ç—å –ø–æ —á–∏—Å–ª—É –æ–±—ä–µ–∫—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ, –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –∏ –ø–æ —É—Ä–æ–≤–Ω—é –¥–æ–≤–µ—Ä–∏—è. –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ —É—Ä–æ–≤–µ–Ω—å –¥–æ–≤–µ—Ä–∏—è 95% –∏ –≤—ã–±–æ—Ä–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç 10 –æ–±—ä–µ–∫—Ç–æ–≤, t* ‚âà 2,262.

–ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞ —Å 9 —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã –≤ —Ç–æ—á–∫–µ t* ‚âà 2,262 —Ä–∞–≤–Ω—è–µ—Ç—Å—è ‚âà 0,975. –¢–æ –µ—Å—Ç—å –º—ã –Ω–µ —É—á–∏—Ç—ã–≤–∞–µ–º 2,5% –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–Ω–æ–π –º–∞—Å—Å—ã –Ω–∞ ¬´–ª–µ–≤–æ–º¬ª —Ö–≤–æ—Å—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∏ 2,5% –Ω–∞ ¬´–ø—Ä–∞–≤–æ–º¬ª.

### –ê–ª–≥–æ—Ä–∏—Ç–º –¥–µ–π—Å—Ç–≤–∏–π –ø–æ–∏—Å–∫–∞ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
–ï—â–µ —Ä–∞–∑ –∑–∞—Ñ–∏–∫—Å–∏—Ä—É–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥–µ–π—Å—Ç–≤–∏–π

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –µ—Å–ª–∏ –Ω–∞–º –∏–∑–≤–µ—Å—Ç–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∫–∞–∫–æ–π-—Ç–æ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –¥–∞–Ω–Ω—ã—Ö –∏ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ, –º—ã –º–æ–∂–µ–º –Ω–∞–π—Ç–∏ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤—Å–µ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö, —Ç.–µ. –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏. –í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ–æ—Ä–º—É–ª–æ–π:

`x—Å—Ä - t* * s/(n**1/2) < Œº < x—Å—Ä + t* * s/(n**1/2)`

–î–ª—è —Ä–µ—à–µ–Ω–∏—è –Ω–∞–º –Ω—É–∂–Ω–æ:

- –≤—ã—á–∏—Å–ª–∏—Ç—å x—Å—Ä - —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –Ω–∞—à–µ–π –≤—ã–±–æ—Ä–∫–µ (–ª–∏–±–æ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –≤ excel, –ª–∏–±–æ –ø–æ —Ñ–æ—Ä–º—É–ª–µ `x—Å—Ä = 1/n*–°–£–ú–ú–êx(i)`, –≥–¥–µ i –æ—Ç 1 –¥–æ n);
- –≤—ã—á–∏—Å–ª–∏—Ç—å `s` - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –≤ –≤—ã–±–æ—Ä–∫–µ (–∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –∏–ª–∏ –ø–æ —Ñ–æ—Ä–º—É–ª–µ `s = ( (–°–£–ú–ú–ê (x(i) - x(—Å—Ä))**2) / n )\*\*1/2`;
- –Ω–∞–π—Ç–∏ –ø–æ —Ç–∞–±–ª–∏—Ü–µ –∑–Ω–∞—á–µ–Ω–∏–µ `t*`: –±–µ—Ä–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ, –≤—ã—á–∏—Ç–∞–µ–º 1, –ø–æ–ª—É—á–∞–µ–º –Ω—É–∂–Ω—É—é —Å—Ç—Ä–æ–∫—É k, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã. –ó–∞—Ç–µ–º –æ–ø—Ä–µ–¥–µ–ª—è–µ–º —Å—Ç–æ–ª–±–µ—Ü —É—Ä–æ–≤–µ–Ω—è –¥–æ–≤–µ—Ä–∏—è (–±–µ—Ä–µ–º 0,9, 0,95 –∏–ª–∏ 0,99 –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∂–µ–ª–∞–µ–º–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏) –∏ –∏—â–µ–º —è—á–µ–π–∫—É –Ω–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–∏. –≠—Ç–æ –∏ –±—É–¥–µ—Ç `t*`;
- –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å `n**1/2`: –∏–∑–≤–ª–µ–∫–∞–µ–º –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –∫–æ—Ä–µ–Ω—å –∏–∑ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ.

–í—Å–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –ø–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –≤ —Ñ–æ—Ä–º—É–ª—É –¥–ª—è `Œº` - –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ –¥–≤–æ–π–Ω–æ–µ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –±—É–¥–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –∏—Å–∫–æ–º—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª.

## 2. –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑
## –ì–∏–ø–æ—Ç–µ–∑—ã
–ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑ ‚Äî —ç—Ç–æ –≤–∞–∂–Ω—ã–π —ç—Ç–∞–ø —Ä–∞–±–æ—Ç—ã –Ω–∞–¥ –ø—Ä–æ–¥—É–∫—Ç–æ–º. –î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑ –Ω—É–∂–Ω–æ –æ—Å–≤–æ–∏—Ç—å —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–π –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–ø–ø–∞—Ä–∞—Ç –∏ –≤–≤–µ—Å—Ç–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π. –í –ø–µ—Ä–≤—É—é –æ—á–µ—Ä–µ–¥—å –Ω–∞–¥–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –¥–≤–µ –≥–∏–ø–æ—Ç–µ–∑—ã, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ –±—É–¥–µ–º —Ä–∞–±–æ—Ç–∞—Ç—å: 

#### –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0) 
- —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ –æ –ø–∞—Ä–∞–º–µ—Ç—Ä–µ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ (–ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π) –∏–ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏, –∫–æ—Ç–æ—Ä–æ–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å. –ù–∞–ø—Ä–∏–º–µ—Ä, –∫–æ–Ω–≤–µ—Ä—Å–∏—è —Å–∞–π—Ç–∞ —Ä–∞–≤–Ω—è–µ—Ç—Å—è 0,7. 

#### –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (HA)

## z-–∫—Ä–∏—Ç–µ—Ä–∏–π
### –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π

#### Z-—Ç–µ—Å—Ç
–ù–∞–ø—Ä–∏–º–µ—Ä, –Ω–∞–º –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –∑–∞—Ä–ø–ª–∞—Ç—ã –∫–ª–∏–µ–Ω—Ç–æ–≤ –°–±–µ—Ä–±–∞–Ω–∫–∞ –∏–º–µ—é—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –∏–∑–≤–µ—Å—Ç–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π `œÉ`. –û–±—ã—á–Ω–æ –∑–∞—Ä–∞–Ω–µ–µ –º—ã –Ω–µ –∑–Ω–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é –≤—Å–µ—Ö –∏–º–µ—é—â–∏—Ö—Å—è –¥–∞–Ω–Ω—ã—Ö (–¥–∏—Å–ø–µ—Ä—Å–∏—é –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏), —ç—Ç–æ—Ç —Å–ª—É—á–∞–π —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –¥–∞–ª–µ–µ.

**—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ (—Ç–æ –µ—Å—Ç—å —Ñ—É–Ω–∫—Ü–∏—è –æ—Ç –∏–º–µ—é—â–∏—Ö—Å—è –¥–∞–Ω–Ω—ã—Ö)**

`Z* = (x—Å—Ä - Œº) / (œÉ / n**1/2)` 

**–∏–º–µ–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å 0 —Å—Ä–µ–¥–Ω–∏–º –∏ –µ–¥–∏–Ω–∏—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π N(0, 1)**.

## –¢–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞
–í –ø—Ä–æ—à–ª–æ–º –ø—Ä–∏–º–µ—Ä–µ –º—ã —Ç—Ä–µ–±–æ–≤–∞–ª–∏, —á—Ç–æ–±—ã –∑–∞—Ä–∞–Ω–µ–µ –±—ã–ª–∞ –∏–∑–≤–µ—Å—Ç–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –∑–∞—Ä–ø–ª–∞—Ç `ùùà`, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è –æ—á–µ–Ω—å —Å–∏–ª—å–Ω—ã–º —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–µ–º –æ—Ç –¥–∞–Ω–Ω—ã—Ö. –ß—Ç–æ–±—ã –æ–±–ª–µ–≥—á–∏—Ç—å —ç—Ç–æ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–µ, –º–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Ç–µ—Å—Ç–æ–º –°—Ç—å—é–¥–µ–Ω—Ç–∞. 

–ü—É—Å—Ç—å –Ω–∞–º –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –∑–∞—Ä–ø–ª–∞—Ç—ã –∫–ª–∏–µ–Ω—Ç–æ–≤ –°–±–µ—Ä–±–∞–Ω–∫–∞ **—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ —Å –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π** `ùùà`, –∏ –Ω–∞–º –Ω—É–∂–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É H0: —Å—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞ —Ä–∞–≤–Ω—è–µ—Ç—Å—è `Œº`. 

–í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –≤–µ—Ä–Ω–æ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ: 

> –µ—Å–ª–∏ –≥–∏–ø–æ—Ç–µ–∑–∞ H0 –≤–µ—Ä–Ω–∞, —Ç–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ `t` –∏–º–µ–µ—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –°—Ç—å—é–¥–µ–Ω—Ç–∞ —Å `k = n-1` —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã:

–ú—ã –ø–æ–º–Ω–∏–º, —á—Ç–æ —ç—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ 0, –∏ —á–µ–º –¥–∞–ª—å—à–µ –æ—Ç 0, —Ç–µ–º –º–∞–ª–æ–≤–µ—Ä–æ—è—Ç–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ **t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏**. –ò–Ω–∞—á–µ –≥–æ–≤–æ—Ä—è, –µ—Å–ª–∏ –º—ã –ø–æ–ª—É—á–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ `t`, –¥–∞–ª–µ–∫–æ–µ –æ—Ç 0, —Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ, –≥–∏–ø–æ—Ç–µ–∑–∞ H0 –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞.

–ù–∞–ø–æ–º–Ω–∏–º, –∑–¥–µ—Å—å `x—Å—Ä` - —Å—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –Ω–∞—à–µ–π –≤—ã–±–æ—Ä–∫–µ –¥–∞–Ω–Ω—ã—Ö, `Œº` - —Å—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞ –ø–æ –≤—Å–µ–º –∫–ª–∏–µ–Ω—Ç–∞–º, `s` ‚Äî —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –ø–æ –≤—ã–±–æ—Ä–∫–µ, –æ–Ω–æ –º–æ–∂–µ—Ç –æ—Ç–ª–∏—á–∞—Ç—å—Å—è –æ—Ç –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–≥–æ —Ä–µ–∞–ª—å–Ω–æ–≥–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è `œÉ`, `n` - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ.

- –µ—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ –∏ –∏–∑–≤–µ—Å—Ç–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å–∏—è, –ø–æ–ª—å–∑—É—é—Ç—Å—è **z-–∫—Ä–∏—Ç–µ—Ä–∏–µ–º**; 

- –µ—Å–ª–∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞ - –ø–æ–ª—å–∑—É—é—Ç—Å—è –∫—Ä–∏—Ç–µ—Ä–∏–µ–º C—Ç—å—é–¥–µ–Ω—Ç–∞.

## –ê–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑
–ò—Ç–∞–∫, —Å–æ–±–µ—Ä–µ–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —à–∞–≥–æ–≤ –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞—à–µ–≥–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –¥–∞–Ω–Ω—ã—Ö.

1. –í [–æ–Ω–ª–∞–π–Ω-–∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä](https://planetcalc.ru/5019/) –≤–≤–æ–¥–∏–º —á–∏—Å–ª–æ —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –∏ –Ω–∞–±–ª—é–¥–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –ø–æ–ª—É—á–∞–µ–º, —á—Ç–æ –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º (—Å–ª–µ–≤–∞ –æ—Ç `t–ù`) –∏–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ, –º–µ–Ω—å—à–µ–µ, —á–µ–º –Ω–∞–±–ª—é–¥–∞–µ–º–æ–µ: `P—Å—Ç(t<tH) = 0,97428`. –ó–Ω–∞—á–µ–Ω–∏–µ –±–æ–ª—å—à–æ–µ, –±–ª–∏–∑–∫–æ–µ –∫ 1, —Ç.–∫. —Å–ª–µ–≤–∞ –æ—Ç –Ω–∞–±–ª—é–¥–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –ø–æ—á—Ç–∏ –≤—Å—è –∫—Ä–∏–≤–∞—è.

2. –¢–æ–≥–¥–∞, —Ç.–∫. –≤—Å—è –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–≤–Ω—è–µ—Ç—Å—è 1, `P—Å—Ç(t>tH) = 1 - P—Å—Ç(t<tH) = 0,02572`. –¢–æ –µ—Å—Ç—å —Å–ø—Ä–∞–≤–∞ –æ—Ç –Ω–∞–±–ª—é–¥–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –æ—Å—Ç–∞–ª—Å—è —Ç–æ–ª—å–∫–æ –º–∞–ª–µ–Ω—å–∫–∏–π "—Ö–≤–æ—Å—Ç–∏–∫" –∫—Ä–∏–≤–æ–π, –ø–æ–¥ –Ω–∏–º –º–∞–ª–µ–Ω—å–∫–∞—è –ø–ª–æ—â–∞–¥—å, –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ, –±–æ–ª—å—à–µ–µ –Ω–∞–±–ª—é–¥–∞–µ–º–æ–≥–æ, –æ—á–µ–Ω—å –º–∞–ª–∞.

–¢–µ–ø–µ—Ä—å –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ–æ—Ä–º—É–ª–æ–π.

3.  `p-value = P—Å—Ç(t>|tH|) = 2 x min{P—Å—Ç(t<tH); P—Å—Ç(t>tH)}`. –ù–∞–º –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ —Å—É–º–º—É –ø–ª–æ—â–∞–¥–µ–π –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º —É —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã—Ö "—Ö–≤–æ—Å—Ç–æ–≤": —Å–ø—Ä–∞–≤–∞ –æ—Ç `t–ù` –∏ —Å–ª–µ–≤–∞ –æ—Ç `-t–ù`. –ó–∞ –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –Ω–∏–º–∏ –æ—Ç–≤–µ—á–∞–µ—Ç –º–µ–Ω—å—à–µ–µ –∏–∑ –Ω–∞–π–¥–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: –±–µ—Ä–µ–º –º–∏–Ω–∏–º—É–º –∏–∑ 0,97428 –∏ 0,02572 ‚Äì —ç—Ç–æ 0,02572 –∏ —É–º–Ω–æ–∂–∞–µ–º –Ω–∞ 2.
