# 1. Теория вероятностей

## Распределения случайных величин
Конечно, не хотелось бы для каждого случайного процесса строить полноценную таблицу. Это не очень удобный формат для восприятия данных. Хотелось бы уметь по любому распределению случайной величины строить удобные и понятные характеристики, которые выражались бы в виде одного-двух чисел. И такие характеристики существуют: это **математическое ожидание** (грубо говоря, среднее значение величины) и **дисперсия** (характерный разброс величины). Более подробно эти понятия будут рассмотрены во втором модуле курса. 

Оказывается, помимо дискретных распределений, которые можно записать в виде таблички, есть еще **непрерывные распределения**. Это зависит от природы случайной величины. Оказывается, бывают: 

- **Дискретные случайные величины** — это те, которые принимают несколько фиксированных значений. Например, оценка пользователем приложений на AppStore. Это случайная величина, которая соответствует либо одной звездочке, либо двум звездочкам, либо трем, либо четырем, либо пяти. То есть всего пять фиксированных значений. В общем случае их может быть сколь угодно много, но главное - счетное количество.

- **Непрерывные случайные величины**, которые принимают произвольные значения или произвольные значения из некоторого диапазона. Например, рост прохожего, стоимость сделки — они могут равняться произвольному действительному числу. Вопрос: а как же задавать непрерывные случайные величины, если множество их значений бесконечно? Выписать их в таблицу не представляется возможным. В этом случае строится график функции f(x), где х — это значения случайной величины, а f(x) — это так называемая плотность вероятности.

Рассмотрим теперь несколько основных случайных распределений, которые наиболее часто описывают реальные жизненные ситуации.

### 1. Равномерное распределение.
Равномерное распределение соответствует случаю, когда все значения случайной величины равновероятны. 

<div><img src="https://cs.sberbank-school.ru/image/full/full/resize/d2734566-9130-11eb-951f-0242ac190003" alt="Пример. Равномерное распределение" style="width: 300px; margin-right: 75%"></div>

Мы с ним уже сталкивались. Например, количество очков на кубике — это равномерное дискретное распределение. Действительно, случайная величина дискретна, она принимает шесть случайных значений — 1, 2, 3, 4, 5 и 6. И вероятность каждого значения 1/6. Соответственно, если мы изобразим это распределение, это будет просто шесть столбиков равной высоты. Все значения случайной величины равновероятны, стало быть, это равномерное распределение. Этот случай соответствует дискретным случайным величинам, но и непрерывные распределения могут быть равномерными (в качестве упражнения постарайтесь придумать пример).

### 2. Нормальное распределение. 
Это непрерывное распределение имеет характерный вид колокола, как показано на графике. У него есть пик в центре, и дальше он симметрично спадает до нуля по мере удаления от пика. 

<div><img src="https://cs.sberbank-school.ru/image/full/full/resize/d54124c0-9130-11eb-b5f5-0242ac190002" alt="Пример. Нормальное распределение" style="width: 400px; margin-right: 75%"></div>

Такое распределение очень часто встречается в жизни, например, разброс при стрельбе. В среднем стрелок всегда попадает в центр, то есть наиболее вероятное значение именно в центре, однако периодически пули попадают левее и правее, и чем дальше попадание пули от центра, тем реже это встречается, потому что в среднем стрелок целится в центр. 

Другой пример — объем инвестиций в стартап. Скажем, в среднем объем инвестиций составляет 10 млн долларов, однако в некоторые стартапы инвестируют 9 млн, в некоторые 11 млн, в некоторые 1 млн, в некоторые 100 млн, но две последних ситуации встречаются гораздо реже, чем среднее значение 10 млн. Например, практически нормальным будет распределение количества сотрудников в различных компаниях. Есть очень большие компании, есть очень маленькие компании, но большинство компаний будет находиться в окрестности некоторого среднего значения. И мы тоже увидим некоторое характерное колоколообразное распределение, очень близкое к нормальному.

### 3. Биномиальное распределение. 
Это дискретное распределение получается в результате так называемой **схемы Бернулли**, имеет похожую на случай нормального распределения форму колокола. Подробнее о схеме Бернулли и примерах можно узнать в видео, которые сюда не подгрузить.

Формула подсчета k успешных сделок из N **независимых** сделок:

- Вероятность успешной сделки: `P(успеха одной сделки) = р`.   
- Вероятность НЕуспешной сделки: `Р(неуспеха одной сделки) = q = 1 - p`.

Задача: посчитать вероятность свершения сценария, в котором k успешных сделок из N.
- `P(k сделок из N окажется удачными)` - ?.

`x(i)` принимает два значения:
- 1, если сделка успешная,
- 0, если сделка неуспешная.

Следовательно, X успешных сделок будет состоять из суммы x(i), в которых учитываться будут только успешные сделки.

Для вывода формулы используется формула подсчета сочетаний, см. тет. по алгебер №1 (Киселев):

`P(k успешных сделок) = C(из N по k) * p\*\*k * (1-p)\*\*(N-k)`

Сочетания: n! / k!\*(N-k)!

Для решения задачи необходимо задать: количество сделок `N`, количество успешных сделок `k`, вероятность успешной сделки `p`.
