```python
# Даны три целых числа. Определите, сколько среди них совпадающих. Программа должна вывести одно из чисел: 
# 3 (если все совпадают), 
# 2 (если два совпадает) или 
# 0 (если все числа различны).

a = input()
b = input()
c = input()

if a == b and b == c:
    print('Result:', 3)
elif (a == b and b != c) or (a != b and b == c) or (a != b and a == c):
    print('Result:', 2)
else:
    print('Result:', 0)
    
```

На данном этапе еще не введено понятие "Функция", поэтому необходимо создать файл и написать код, который будет похож на примеры из раздела "Циклы". Сначала необходимо объявить переменные, затем написать код, удовлетворяющий условию задания. Для вывода результата необходимо использовать функцию print().

Разрешается использовать функцию str(число). Данная функция переводит числовые данные в строковые.

```python
# 5. Дополните код из предыдущего задания, чтобы теперь получался ромб. 
# То есть количество ступеней должно равняться числу n*2-1.

# n = int(input('Enter the number between 1 and 9: '))
n = 5
row1 = '1'
row2 = ''

if n > 0 and n <= 9:
    
    print(' ' * (n - 1), '1')          # opening line shows '1'
    
    for i in range(2, n + 1):          # creating the first half of 'rhombus'
        row1 = row1 + str(i)
        row2 = str(i - 1) + row2
        print(' ' * (n - i), row1 + row2)
        
    for i in range(1, n - 1):          # creating the second half of the 'rhombus' 
        i = n - i                      # the count goes back from the biggest number to the lowest
        row1 = '1'
        row2 = ''
        for j in range(2, i + 1):      # the internal loop for creating backwards lines
            row1 = row1 + str(j)
            row2 = str(j - 1) + row2
        print(' ' * (n - i), row1 + row2)   
    
    print(' ' * (n - 1), '1')          # closing line repeats the opening line

else:
    print('Something went wrong...')
```

### Строки. Задание 1
Дана строка: Abraсadabra

Требуется:

1. Вывести третий символ этой строки.
1. Вывести предпоследний символ этой строки.
1. Вывести первые пять символов этой строки.
1. Вывести строку, кроме последних двух символов.
1. Вывести все символы с четными индексами (считайте, что 0 - четный индекс).
1. Вывести все символы с нечетными индексами.
1. Вывести все символы в обратном порядке.
1. Вывести все символы строки через один в обратном порядке, начиная с последнего.
1. Вывести длину данной строки.

```python
a = 'Abracadabra'

print('Given string:', a)

# 1. Вывести третий символ этой строки.
print('1. Third symbol:', a[2])

# 2. Вывести предпоследний символ этой строки.
print('2. Second symbol from the end:', a[-2])

# 3. Вывести первые пять символов этой строки.
print('3. First 5 symbols:', a[0:5])

# 4. Вывести строку, кроме последних двух символов.
print('4. The string w/out two symbols in the end:', a[0 : len(a) - 2])

# 5. Вывести все символы с четными индексами (считайте, что 0 - четный индекс).
evens = ''
for count in range(0, len(a)):
    if count % 2 == 0:
        evens = evens + a[count]
print('5. Even-numbered symbols:', evens)

# 6. Вывести все символы с нечетными индексами.
odds = ''
for count in range(0, len(a)):
    if count % 2 == 1:
        odds = odds + a[count]
print('6. Odd numbered symbols:', odds)

# 7. Вывести все символы в обратном порядке.
backwards = ''
for count in range(0, len(a)):
    count = len(a) - 1 - count
    backwards = backwards + a[count]
print('7. Backwards:', backwards)

# 8. Вывести все символы строки через один в обратном порядке, начиная с последнего.
# (доработано для любой строки)
bwards = ''
for count in range(0, len(a)):
    count = len(a) - 1 - count
    if len(a) % 2 == 0:
        if count % 2 == 1:
            bwards = bwards + a[count]        
    elif len(a) % 2 == 1:
        if count % 2 == 0:
            bwards = bwards + a[count]        
print('8. Backwards thinned:', bwards)
            
# 9. Вывести длину данной строки.
print('9. And the length is:', len(a))
```

### Строки. Задание 2
Напишите код, который обрабатывает строковые данные и возвращает их с первыми заглавными буквами в каждом слове.

#### Примечание
Слова в строке отделены пробелами. Каждое слово после обработки должно быть с заглавной буквы, только, если первый символ слова - буква, в случае типа '3amg' код не должен менять буквы.

На данном этапе еще не введено понятие "Функция", поэтому необходимо создать файл и написать код, как в предыдущих разделах. Сначала необходимо объявить переменные, затем написать код, удовлетворяющий условию задания. Для вывода результата необходимо использовать функцию print().

```python
string = 'a1 2b  3   abc d3e r2D2'
new_string = ''

for word in string.split():
    if word[0].isalpha():
        letter = word[0].upper()
        for i in range(len(word)):
            if i == 0:
                continue
            else:
                letter = letter + word[i]
        new_string = new_string + letter
    else:
        new_string = new_string + word

final_string = ''
count = 0
for letter in new_string:    
    while letter.lower() != string[count].lower():
        final_string = final_string + ' '  
        count += 1
    final_string = final_string + letter
    count += 1

print(string)
print(final_string)
    
```

```python
numbers = '1234567890'
lletters = 'abcdefghijklmnopqrstuvwxyz'
uletters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
signs = '!@#$%^&*()-+'

print('Придумайте пароль из 12 символов или больше.')
print('Пароль должен состоять из чисел, строчных и заглавных букв, а также следующих спецсимволов: ', end='')
print(signs, '.')
pswd = input('\nВведите пароль: ')

numbers_check = 0
lletters_check = 0
uletters_check = 0
signs_check = 0
error = 0
check = ((numbers_check != 0) 
    or (lletters_check != 0) 
    or (uletters_check != 0) 
    or (signs_check != 0) 
    or (len(pswd) >= 12))

for symbol in pswd:        
    if symbol in numbers: numbers_check += 1            
    elif symbol in lletters: lletters_check += 1        
    elif symbol in uletters: uletters_check += 1        
    elif symbol in signs: signs_check += 1
    else:
        error = 1
        break
        
if error == 1:
    print('Ошибка. Запрещенный спецсимвол.')
elif ((numbers_check != 0) 
    and (lletters_check != 0) 
    and (uletters_check != 0) 
    and (signs_check != 0) 
    and (len(pswd) >= 12)):
    print('Сильный пароль.')
else:
    recommendations = 'Слабый пароль. Рекомендации: '
    if len(pswd) < 13: recommendations = recommendations + 'увеличить число символов на ' + str(13 - len(pswd))
    if (numbers_check == 0) and (check == 0): recommendations = recommendations + ', добавить число'
    if (numbers_check == 0) and (check == 1): recommendations = recommendations + 'добавить число'
    if (lletters_check == 0) and (check == 0): recommendations = recommendations + ', добавить строчную букву'
    if (lletters_check == 0) and (check == 1): recommendations = recommendations + 'добавить строчную букву'
    if (uletters_check) == 0 and (check == 0): recommendations += ', добавить заглавную букву'
    if (uletters_check) == 0 and (check == 1): recommendations += 'добавить заглавную букву'
    if (signs_check == 0) and (check == 0): recommendations = recommendations + ', добавить спецсимвол'
    if (signs_check == 0) and (check == 1): recommendations = recommendations + 'добавить спецсимвол'
    print(recommendations + '.')
```

## 15. Работа с файлами. Задания

Задание 1
В данной задаче вам будет необходимо работать с заранее подготовленным файлом, представленным ниже. Текстовый файл находится в ZIP архиве, вам необходимо скачать его, распаковать и переместить текстовый документ в папку, где будет лежать файл с программой для решения данной задачи. 
Файл представляет собой список данных людей со следующими заголовками:
    
EMAIL NAME LAST_NAME TEL CITY

Столбец EMAIL пустой у всех. Остальные позиции могут быть заполнены или нет.
Представим следующую ситуацию: вы являетесь сотрудником IT отдела некой компании. Ваша компания открывает новый филиал с переводом необходимого количества сотрудников из разных городов. Новый филиал имеет свой почтовый домен. Вам, как сотруднику IT отдела, необходимо создать каждому сотруднику, переходящему в новый филиал, свой уникальный почтовый адрес. У вас есть неотформатированный список сотрудников, который заполнялиб не понятно как. Также у вас есть кусок кода - функция, которая по имени и фамилии сотрудника составляет его уникальный почтовый адрес. Для выполнения данного задания вам необходимо написать программу, удовлетворяющую следующим требованиям:

1. Программа должна читать исходный текстовый файл.

2. Программа должна содержать функцию, представленную ниже, для создания почтовых адресов.

3. Программа должна заполнить пустой столбец с почтовыми адресами в исходном файле и перезаписать его.

4. Почтовый адрес должен создаваться только для тех, у кого заполнены все остальные поля - Имя, Фамилия, Телефонный номер, Город.

5. Телефонный номер считается валидным, если он состоит из 7 цифр, в противном случае информация невалидна и программа не должна создавать почтовый адрес сотруднику.

После выполнения вашей программы исходный текстовый файл должен быть заполнен информацией с почтовыми адресами сотрудников. 

Функция для создания почтовых адресов:

```python
def email_gen(list_of_names):
    emails = []
    for i in list_of_names:
        letter = 1
        while i[1] + '.' + i[0][0:letter] + '@company.io' in emails:
            letter+=1
        emails.append(i[1] + '.' + i[0][0:letter] + '@company.io')
    return emails
```

Функция принимает список списков **[['Имя_1', 'Фамилия_1'], ['Имя_2', 'Фамилия_2'], ['Имя_3', 'Фамилия_3']]** из имен и фамилий **list_of_names**. Далее создается пустой список emails, который в ходе выполнения, заполнится почтовыми адресами. Запускается цикл для каждого списка из имени и фамилии **['Имя_1', 'Фамилия_1']**. Устанавливается окончание среза **letter = 1**. Далее проходит проверка на совпадение, если в списке адресов уже есть такой же адрес (например имя и фамилия сотрудников совпадают), то срез i[0][0:letter] увеличивается на 1 символ. После в список адресов добавляется новый адрес. Адреса присваиваются пользователям по следующей формуле: берется фамилия сотрудника, добавляется точка, затем добавляется первая буква имени и остаточная часть '@company.io'. Если имена и фамилии сотрудников совпадают, к первой букве имени добавляется вторая и т.д.

Пример выполнения:

```python
print(email_gen([['Ivan', 'Petrov'], ['Ivan', 'Petrov'], ['Ivan', 'Petrov']]))
['Petrov.I@company.io', 'Petrov.Iv@company.io', 'Petrov.Iva@company.io']
```

При обработке файла обращайте внимание на пробелы. Почтовые адреса не должны их содержать. 

Входным параметром для программы является приложенный файл.
Выходным значением должен быть файл с заполнеными почтовыми адресами.

После получения итогового файла изучите его, обратите внимание на заполненные поля невалидными данными. Подумайте о том, как можно еще доработать программу, чтобы избежать обработки невалидных данных.

```python
def email_gen(list_of_names):
    emails = []
    for i in list_of_names:
        letter = 1
        while i[1] + '.' + i[0][0:letter] + '@company.io' in emails:
            letter+=1
        emails.append(i[1] + '.' + i[0][0:letter] + '@company.io')
    return emails

fhand = open('task_file.txt', 'r+')

lst = []
bad_lines = 0
for line in fhand:
    line = line.split()
    if ((len(line) < 5)
        or (len(line[1]) <= 2)
        or (len(line[2]) <= 2)
        or ('NO_NAME,' in line)
        or (line[1].islower() == True)):
        bad_lines += 1
        print('Bad line:', line)
    else:
        line.pop()
        line.pop()
        line.pop(0)
        count = 0
        for i in line:            
            i = i.strip(',')
            line[count] = i
            count += 1
        lst.append(line)
#         print(line)
lst.pop(0)

# print(lst)
print('Bad lines total:', bad_lines)
print('Correct lines total:', len(lst))

print('\n', email_gen(lst))
```
