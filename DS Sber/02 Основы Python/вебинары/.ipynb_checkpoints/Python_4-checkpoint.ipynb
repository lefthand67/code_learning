–¢–∞–∫–∂–µ –≤—ã –º–æ–∂–µ—Ç–µ —Å—Ç–æ–ª–∫–Ω—É—Ç—å—Å—è —Å –ø—Ä–æ–±–ª–µ–º–æ–π –∫–æ–¥–∏—Ä–æ–≤–∫–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Ñ–∞–π–ª–æ–≤. –ï—Å–ª–∏ –≤–º–µ—Å—Ç–æ —Ä—É—Å—Å–∫–∏—Ö –±—É–∫–≤, –≤—ã –≤–∏–¥–∏—Ç–µ –∫—Ä–∞–∫–æ–∑—è–±—Ä—ã, —Ç–æ —É–∫–∞–∂–∏—Ç–µ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ `open()` –≤—Ç–æ—Ä–æ–π –∞—Ä–≥—É–º–µ–Ω—Ç - —Ç–∏–ø –∫–æ–¥–∏—Ä–æ–≤–∫–∏ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ —Ñ–∞–π–ª–∞:

–ú–µ—Ç–æ–¥ `.read()` —Ç–∞–∫–∂–µ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ–¥–∏–Ω –∞—Ä–≥—É–º–µ–Ω—Ç `size`, –∫–æ—Ç–æ—Ä—ã–π –æ—Ç–≤–µ—á–∞–µ—Ç –∑–∞ —Ç–æ, —Å–∫–æ–ª—å–∫–æ —Å–∏–º–≤–æ–ª–æ–≤ –∏–∑ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ —Ñ–∞–π–ª–∞ –±—É–¥–µ—Ç –≤—ã–≤–µ–¥–µ–Ω–æ –Ω–∞ —ç–∫—Ä–∞–Ω. –û–¥–Ω–∞–∫–æ, –µ—Å–ª–∏ –≤—ã–∑–æ–≤–µ—Ç–µ —ç—Ç–æ—Ç –º–µ—Ç–æ–¥ –µ—â–µ —Ä–∞–∑ —Å –¥–∞–Ω–Ω—ã–º –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º, —Ç–æ Python –ø—Ä–æ–¥–æ–ª–∂–∏—Ç –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ —Ä–æ–≤–Ω–æ —Å —Ç–æ–≥–æ –º–µ—Å—Ç–∞, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º –æ–Ω –æ—Å—Ç–∞–Ω–æ–≤–∏–ª—Å—è –Ω–∞ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏. –¢–æ –µ—Å—Ç—å —ç—Ç–æ –∞–Ω–∞–ª–æ–≥ –ø–∏—à—É—â–µ–π –∫–∞—Ä–µ—Ç–∫–∏. –ü–æ—ç—Ç–æ–º—É, —á—Ç–æ–±—ã –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –µ—â–µ —Ä–∞–∑, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –º–µ—Ç–æ–¥–æ–º `.seek()` –∏ –ø–µ—Ä–µ—Å—Ç–∞–≤–∏—Ç—å —ç—Ç—É –∫–∞—Ä–µ—Ç–∫—É –≤ –Ω–∞—á–∞–ª–æ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ —Ñ–∞–π–ª–∞. –ú–µ—Ç–æ–¥ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∞—Ä–≥—É–º–µ–Ω—Ç - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–∞–π—Ç –æ—Ç –Ω–∞—á–∞–ª–∞ —Ñ–∞–π–ª–∞. –£—á—Ç–∏—Ç–µ, —á—Ç–æ –≤—Å–µ —Å–∏–º–≤–æ–ª—ã –≤ –∫–æ–¥–∏—Ä–æ–≤–∫–∞—Ö –º–æ–≥—É—Ç –∑–∞–Ω–∏–º–∞—Ç—å —Ä–∞–∑–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞–º—è—Ç–∏. –¢–∞–∫ –Ω–∞–ø—Ä–∏–º–µ—Ä, –∞–Ω–≥–ª–∏–π—Å–∫–∏–µ –±—É–∫–≤—ã –≤ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–µ —Å–≤–æ–µ–º –∑–∞–Ω–∏–º–∞—é—Ç 8 –±–∏—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –∫–∞–∂–¥–∞—è. –ê –±—É–∫–≤—ã —Ä—É—Å—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞ –±—É–¥—É—Ç –∑–∞–Ω–∏–º–∞—Ç—å 16 –±–∏—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏. –¢–æ –µ—Å—Ç—å –¥–≤–∞ –±–∞–π—Ç–∞. –ú–µ—Ç–æ–¥ seek –ø–µ—Ä–µ—Å—Ç–∞–≤–ª—è–µ—Ç –ø–æ–±–∞–π—Ç–æ–≤–æ, –ø–æ—ç—Ç–æ–º—É –¥–ª—è –ø–µ—Ä–µ–º–µ—â–µ–Ω–∏—è –ø–æ –±—É–∫–≤–∞–º —Ä—É—Å—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞ –Ω—É–∂–Ω–æ —ç—Ç–æ —á–∏—Å–ª–æ —É–º–Ω–æ–∂–∞—Ç—å –Ω–∞ 2 –∏ —Ç–æ–≥–¥–∞ –ø–æ–ª—É—á–∏—Ç—Å—è –ø–µ—Ä–µ–º–µ—â–∞—Ç—å—Å—è –ø–æ —Ç–µ–∫—Å—Ç—É –∏–∑ —Ä—É—Å—Å–∫–∏—Ö –±—É–∫–≤. –û–¥–Ω–∞–∫–æ, –ø—Ä–æ–±–µ–ª—ã –∑–∞–Ω–∏–º–∞—é—Ç —Ç–æ–∂–µ –ø–æ –æ–¥–Ω–æ–º—É –±–∞–π—Ç—É.

### –ó–∞–¥–∞—á–∞ 11. –°–æ–∑–¥–∞–Ω–∏–µ —Ñ–∞–π–ª–æ–≤
–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `create_file_with_numbers`, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –æ–¥–Ω–æ —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ - `n`.

–§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ —Å–æ–∑–¥–∞—Ç—å —Ñ–∞–π–ª —Å –Ω–∞–∑–≤–∞–Ω–∏–µ "range_[number].txt" –∏ –Ω–∞–ø–æ–ª–Ω–∏—Ç—å –µ–≥–æ —Ü–µ–ª—ã–º–∏ —á–∏—Å–ª–∞–º–∏ –æ—Ç 1 –¥–æ n –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ, –ø—Ä–∏—á–µ–º –∫–∞–∂–¥–æ–µ —á–∏—Å–ª–æ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è  –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ. –ù–µ –∑–∞–±—É–¥—å—Ç–µ —É–∫–∞–∑–∞—Ç—å –ø–æ–ª–Ω—ã–π –ø—É—Å—Ç—å –∫ –≤–∞—à–µ–π –ø–∞–ø–∫–µ –Ω–∞ Google Drive.

–ü—Ä–∏–º–µ—Ä: —Ñ—É–Ω–∫—Ü–∏—è `create_file_with_numbers(5)` –¥–æ–ª–∂–Ω–∞ —Å–æ–∑–¥–∞—Ç—å —Ñ–∞–π–ª "range_5.txt" —Å —Å–æ–¥–µ—Ä–∂–∏–º—ã–º


```
1
2
3
4
5
```




````python
# –í–∞—à–µ —Ä–µ—à–µ–Ω–∏–µ

def create_file_with_numbers(n: int) -> None:
    fhand = open(f'range_{n}.txt', mode='w')
    for i in range(1, n+1):
        fhand.write(str(i) + '\n')
    fhand.close()

create_file_with_numbers(6)
````

### –ó–∞–¥–∞—á–∞ 12*. –ü–æ–∏—Å–∫ —Å–∞–º–æ–≥–æ –¥–ª–∏–Ω–Ω–æ–≥–æ —Å–ª–æ–≤–∞ –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º —Ñ–∞–π–ª–µ

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `longest_word_in_file`, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∏–º—è —Ñ–∞–π–ª–∞ –∏ –≤–Ω—É—Ç—Ä–∏ –µ–≥–æ —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ –Ω–∞—Ö–æ–¥–∏—Ç —Å–∞–º–æ–µ –¥–ª–∏–Ω–Ω–æ–µ —Å–ª–æ–≤–æ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –µ–≥–æ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –æ—Ç–≤–µ—Ç–∞. –í —Å–ª—É—á–∞–µ,  –µ—Å–ª–∏ –µ—Å—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–ª–æ–≤ —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –¥–ª–∏–Ω–æ–π, –Ω—É–∂–Ω–æ –≤–µ—Ä–Ω—É—Ç—å —Å–ª–æ–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–Ω–µ–µ –≤ —Ç–µ–∫—Å—Ç–µ.

–ü—Ä–∏ —ç—Ç–æ–º —Å–ª–æ–≤–∞ –≤ —Ç–µ–∫—Å—Ç–µ –æ—Ç–¥–µ–ª—è—é—Ç—Å—è –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –ø—Ä–æ–±–µ–ª–∞–º–∏, –ª—é–±—ã–µ –¥—Ä—É–≥–∏–µ –∑–Ω–∞–∫–∏ –ø—É–Ω–∫—Ç—É–∞—Ü–∏–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–∫–ª—é—á–∏—Ç—å.

–í—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –∑–Ω–∞–∫–∏ –ø—É–Ω–∫—Ç—É–∞—Ü–∏–∏ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å –∏–∑ –º–æ–¥—É–ª—è `string`

```
from string import punctuation
```



````python
# –í–∞—à–µ —Ä–µ—à–µ–Ω–∏–µ

def longest_word_in_file():
    from string import punctuation
    file_name = input('File name: ')
    fhand = open(file_name, 'r', encoding='utf-8')
    longest_word = None
    for line in fhand:
        words = line.split()
        for word in words:
            word = word.rstrip(punctuation)
            if longest_word is None or len(word)>=len(longest_word):
                longest_word = word
    fhand.close()
    return print(longest_word)

longest_word_in_file()
````
